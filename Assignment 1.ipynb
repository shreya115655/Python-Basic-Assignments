10. Why does this expression cause an error? How can you fix it?
'I have eaten ' + 99 + ' burritos.'
Ans:This expression causes and error because in this line 'I have eaten' and 'burritos' are strings, while 99 is an integer. In order to fix the error and print 'I have eaten 99 burritos.', 99 needs '' around it to treat it as a string . 
So, the corrected one will be :'I have eaten ' + '99' + ' burritos.'--> 'I have eaten 99 burritos.'
